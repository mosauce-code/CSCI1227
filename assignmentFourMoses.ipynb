# **Q1. Calculating Average Compressive Strength (MPa) of Concrete**

*Important: You cannot use the **sum()** or **len()** functions in your Python
code. You should try to replicate their functionality using what you have learnt
about for / while loops.*

```python
# Don't change the code below ðŸ‘‡
data = input(
    "Enter a number of concrete compresive strength values (MPa) separated by space: ").split()
for i in range(0, len(data)):
    data[i] = float(data[i].strip())
# Don't change the code above ðŸ‘†

# Write your code below this row ðŸ‘‡

sum = 0
count = 0
for value in data:
  sum += value #Adds, like a+b+c+d+e........
  count += 1 #Iteration count.

avg = round((sum / count), 2) #Using the avg formula using round().
print(f"The average compressive strength of the concrete is {avg} MPa.")
```

# **Q2. SimpleATM**
### **ATM Program Steps**

#### **1. Print a Welcome Message**  
- Display a welcome message when the program starts.

#### **2. Ask for PIN & Validate (3 Attempts)**  
- Prompt the user to enter their PIN.  
- If the entered PIN is incorrect, notify the user and show remaining attempts.  
- Allow up to **3 attempts** to enter the correct PIN.  
- If the user fails all attempts, display a goodbye message and exit the program.  
- If the correct PIN is entered, proceed to the next step.  

#### **3. Display Main Menu**  
- Show the following options to the user:  
  1. Change PIN  
  2. View Balance  
  3. Withdraw Money  
  4. Deposit Money  
  5. Exit  

- Ask the user to select an option.  
- If an invalid option is entered, notify the user and show the menu again.  

#### **4. Change PIN**  
- Ask for the **current PIN** to verify identity.  
- If incorrect, allow **3 attempts** to enter the correct PIN.  
- If correct, prompt the user to enter a new **4-digit PIN**.  
- If the new PIN is valid, update it and show a success message.  
- If invalid or incorrect attempts exceed **3**, show an error message and return to the menu.  

#### **5. View Balance**  
- Display the current account balance.  
- Return to the main menu.  

#### **6. Withdraw Money**  
- Ask the user to enter the withdrawal amount.  
- Check if there are **sufficient funds** in the balance:  
  - If **yes**, deduct the amount and show the updated balance.  
  - If **no**, display an insufficient funds message.  
- Return to the main menu.  

#### **7. Deposit Money**  
- Ask the user to enter the deposit amount.  
- Add the amount to the balance and display the updated total.  
- Return to the main menu.  

#### **8. Exit the Program**  
- If the user selects **Exit (5)**, print a goodbye message and terminate the program.

```python
#Coding the ATM again, using defined chunks of code, using try and except.

PIN = "3741" #Used random number generators for these numbers
balance = 10910

# you should not change the following lines of code
menuOptions = """Menu Options:
            1. Set a new PIN
            2. View Balance
            3. Withdraw
            4. Deposit
            5. Exit

        """

bye = "Thanks for using mosaucebank. Goodbye!"

# UPDATE - change your name in the welcome message at the terminal
print("Welcome to Moses' mosaucebank! \n")
def _main(PIN): #Main function, to check if the initial PIN is correct.
  attempts = 3
  while attempts > 0:
    pin = input("Enter your PIN: ")
    if pin == PIN:
      print("Access granted. Welcome to mosaucebank!")
      attempts = 0
      return True
    else:
      print(f"Access denied. You have {attempts-1} attempt(s) left. ")
      attempts -= 1
      if attempts == 0:
        print("Too many incorrect attempts. Terminating program. ")
        break


def _change_pin(PIN):
    attempts = 3
    while attempts >= 0:
        old_pin = input("Enter your current PIN: ")
        if old_pin == PIN:
          new_pin = input("Enter a new 4-digit PIN: ")

          while len(new_pin) != 4 or not new_pin.isdigit():
            print("Invalid PIN! Please enter a 4-digit number. ")
            new_pin = input("Enter a new 4-digit PIN: ")

          PIN = new_pin
          print("PIN changed successfully. ")
          return PIN

        else:
          print(f"Incorrect PIN. You have {attempts-1} attempt(s) left. ")
          attempts -= 1
          if attempts == 0:
            print("Too many incorrect attempts. Redirecting to menu.  ")
            return PIN






def _view_balance(balance):
    print(f"Your current balance is ${balance}.")

def _withdraw(balance):
  try:
      withdraw_amount = int(input("How much would you like to withdraw?"))
      if withdraw_amount > balance:
        print("Insufficient funds.")
      else:
        balance -= withdraw_amount
        print(f"Your new balance is ${balance}.")
  except ValueError:
    print("Invalid input. Please enter a valid number.")
  return balance

def _deposit(balance):
  try:
    deposit_amount = int(input("How much would you like to deposit?"))
    balance += deposit_amount
    print(f"Your new balance is ${balance}.")
  except ValueError:
    print("Invalid input. Please enter a valid number.")
  return balance


if _main(PIN):
  while True:
    print(menuOptions)
    choice = input("Enter your choice: ")

    match choice:
      case "1":
        PIN = _change_pin(PIN)
      case "2":
        _view_balance(balance)
      case "3":
        balance = _withdraw(balance)
      case "4":
        balance = _deposit(balance)
      case "5":
        print(bye)
        break
      case _:
        print("Invalid choice. Please try again.")

```
